This project uses `Sales and Trends Dataset`.

- TransactionID (Unique identifier for each transaction)
- Date (Date of the transaction, format: YYYY-MM-DD)
- Time (Time of the transaction, format: HH:MM:SS)
- CustomerID (Unique identifier for each customer)
- Age (Age of the customer)
- Gender (Gender of the customer: Male, Female, Other)
- Location (City or town where the purchase was made)
- StoreID (Unique identifier for the store, if applicable)
- OnlineOrderFlag (Boolean: True if online, False if in-store)
- ProductID (Unique identifier for the product)
- ProductName (Name of the product)
- Category (Category of the product, e.g., Electronics, Clothing, Toys, Food, Decorations)
- Quantity (Number of items purchased in the transaction)
- UnitPrice (Price per unit of the product)
- TotalPrice (Total price for the product, calculated as Quantity * UnitPrice)
- PaymentType (Type of payment, e.g., Credit Card, Debit Card, Cash, Online Payment)
- Promotion Applied (Boolean: True if any promotion was applied, False otherwise)
- DiscountAmount (The amount of discount, if any)
- GiftWrap (Boolean: True if the product was gift-wrapped, False otherwise)
- ShippingMethod (Method of shipping, e.g., Standard, Express, Overnight, if online)
- Delivery Time (Number of days taken for delivery, if online)
- Weather (General weather condition on the day of purchase, e.g., Snowy, Rainy, Sunny)
- Event (Special events on the purchase day, e.g., Christmas Market, Black Friday)
- Customer Satisfaction (Customer satisfaction rating, on a scale of 1-5)
- Return Flag (Boolean: True if the product was returned, False otherwise)

- ### Date Column:

```python
Sales['Date'] = Sales['Date'] .astype('str')
```

```python
Sales.insert(2,'Year',0)
Sales['Year'] = Sales['Date'].str[:4]
Sales['Year'] = Sales['Year'].astype('int64')
```

```python
Sales.insert(3,'Month',0)
Sales['Month'] = Sales['Date'].str[5:7]
Sales['Month'] = Sales['Month'].astype('int64')
```

```python
Sales.insert(4,'Day',0)
Sales['Day'] = Sales['Date'].str[8:]
Sales['Day'] = Sales['Day'].astype('int64')
```

```python
Sales['Month'] = Sales['Month'].replace({1:'Jan' ,2:'Feb' ,3:'Mar', 4:'Apr',5:'May',6:'June',7:'July',8:'Aug',9:'Sept',10:'Oct',11:'Nov',12:'Dec'})
```

```python
Sales.head().style.background_gradient(cmap='Blues')
```

```python
Sales = Sales.drop(labels =
   'Date',
   axis=1 )
```

- ### Time column:

```python
Sales['Time'] = Sales['Time'] .astype('str')
```

```python
Sales.insert(5,'Hour',0)
Sales['Hour'] = Sales['Time'].str[:2]
Sales['Hour'] = Sales['Hour'].astype('int64')
```

```python
Sales.head().style.background_gradient(cmap='Blues')
```

```python
Sales.insert(6,'Daytime','')
Sales.loc[(Sales['Hour'] >= 6) & (Sales['Hour'] <= 11),'Daytime'] = 'Morning'
Sales.loc[(Sales['Hour'] >= 12) & (Sales['Hour'] <= 17),'Daytime'] = 'Afternoon'
Sales.loc[(Sales['Hour'] >= 18) & (Sales['Hour'] <= 21),'Daytime'] = 'Evening'
Sales.loc[(Sales['Hour'] >= 22) | (Sales['Hour'] <= 5),'Daytime'] = 'Night'
```

```python
Sales.head().style.background_gradient(cmap='Blues')
```

```python
Sales = Sales.drop(labels =
   'Time',
   axis=1 )
```

```python
Sales.head().style.background_gradient(cmap='Blues')
```

- ## Product Sales Distribution

```python
product_counts = Sales['ProductName'].value_counts()

# Create a blue color palette
num_products = len(product_counts)
blue_colors = plt.cm.Blues([i / num_products for i in range(num_products)])  # Generate a list of blue colors

# Create a pie chart for product distribution
plt.figure(figsize=(8, 8))

# Plot the pie chart
plt.pie(product_counts, labels=product_counts.index, autopct='%1.1f%%', startangle=90, colors=blue_colors)

# Add a title
plt.title('Product Distribution', fontsize=18)

# Ensure equal aspect ratio so the pie is circular
plt.axis('equal')

# Display the plot
plt.show()
```

- ## Monthly Sales Over Time

```python

# Ensure the 'Month' column is numeric
Sales['Month'] = Sales['Month'].apply(lambda x: pd.to_datetime(x, format='%b').month if isinstance(x, str) else x)

# Count monthly sales
monthly_sales = Sales.groupby(['Year', 'Month'])['TotalPrice'].sum().reset_index()

# Create a date for plotting
monthly_sales['Date'] = pd.to_datetime(monthly_sales[['Year', 'Month']].assign(DAY=1))

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Date'], monthly_sales['TotalPrice'], marker='o', color='blue')
plt.title('Monthly Sales Over Time', fontsize=18)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total Sales ($)', fontsize=14)
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

- ## Total sales by product

```python

product_sales = Sales.groupby('ProductName')['TotalPrice'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
product_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Product', fontsize=18)
plt.xlabel('Product', fontsize=14)
plt.ylabel('Total Sales ($)', fontsize=14)
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

- ## Count the occurrences of each satisfaction level

```python

satisfaction_counts = Sales['CustomerSatisfaction'].value_counts()

# Define a blue color palette
colors = ['#1f77b4', '#aec7e8', '#ff7f0e', '#ffbb78']  # You can adjust the colors here

plt.figure(figsize=(8, 8))
plt.pie(satisfaction_counts, labels=satisfaction_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Customer Satisfaction Distribution', fontsize=18)
plt.axis('equal')
plt.show()
```

- ## Total sales by gender

```python

gender_sales = Sales.groupby('Gender')['TotalPrice'].sum()

# Create the bar chart with specified colors
plt.figure(figsize=(8, 6))
gender_sales.plot(kind='bar', color=['pink', 'blue', 'gray'])  # Update to the specified colors
plt.title('Total Sales by Gender', fontsize=18)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Total Sales ($)', fontsize=14)
plt.xticks(rotation=0)
plt.grid(axis='y')  # Only show grid lines for the y-axis
plt.tight_layout()
plt.show()
```

- ## Total sales by age bracket

```python

age_sales = Sales.groupby('Age Bracket')['TotalPrice'].sum()

# Create a blue color palette with enough colors
num_colors = len(age_sales)
blue_colors = plt.cm.Blues([i / num_colors for i in range(num_colors)])  # Generates shades of blue

# Create the bar chart with the blue palette
plt.figure(figsize=(10, 6))
bars = plt.bar(age_sales.index, age_sales, color=blue_colors)  # Apply the blue colors

# Adding color to each bar explicitly (optional for debugging)
for bar in bars:
    bar.set_color(blue_colors[bars.index(bar) % num_colors])

plt.title('Total Sales by Age Bracket', fontsize=18)
plt.xlabel('Age Bracket', fontsize=14)
plt.ylabel('Total Sales ($)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y')  # Only show grid lines for the y-axis
plt.tight_layout()
plt.show()

```
