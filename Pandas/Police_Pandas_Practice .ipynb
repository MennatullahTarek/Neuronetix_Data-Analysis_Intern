# <p style="padding:10px;background-color:#f6cd04;margin:0;color:#0c4da3;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 50px 50px;overflow:hidden;font-weight:200">II. Missing Data ‚ùî‚ùî</p> 



```python
def missing_values_table(df):
  mis_val = df.isnull().sum()
  mis_val_percent = 100*df.isnull().sum() / len(df)
  mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
  mis_val_table_ren_columns = mis_val_table.rename(columns = {0:'Missing Values' , 1:'of Total Values %'})
  return mis_val_table_ren_columns.round(2)
```

```python
print("Missing In Police Data BEFORE :- ")
missing_values_table(police).style.background_gradient(cmap='Blues')
```

- #### <span style="color:#08306B;"> From the previous table `'search_type'` Column has more than `50% null values`, So let's `drop` it. </span>

```python
police.drop('search_type', axis=1 , inplace=True)
```

- #### <span style="color:#08306B;">  There are about `5123` Nan rows, So let's `drop` it.</span>

```python
police.dropna(subset = ['driver_race'], inplace=True)
```

```python
print("Missing In Police Data :- ")
missing_values_table(police).style.background_gradient(cmap='Blues')
```

- #### <span style="color:#08306B;">  `'driver_age'` and `'driver_gender'` columns also have null values, So how we can deal with them. </span>

```python
police['driver_age'].unique()
```

```python
mean_age = police['driver_age'].mean()
```

```python
police['driver_age'] = police['driver_age'].fillna(mean_age)
```

```python
police['driver_gender'].unique()
```

```python
police.dropna(subset = ['driver_gender','driver_age_raw'], inplace=True)
```

```python
print("Missing In Police Data After :- ")
missing_values_table(police).style.background_gradient(cmap='Blues')
```

- #### <span style="color:#08306B;">  Now our data is free of missing values `:)` </span>

<a class="anchor" id="correction"></a>
# <p style="padding:10px;background-color:#f6cd04;margin:0;color:#0c4da3;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 50px 50px;overflow:hidden;font-weight:200">III. Data Correction ‚öôÔ∏è</p> 



```python
police.insert(1,'stop_year',0)
police['stop_year'] = police['stop_date'].str[:4]
police['stop_year'] = police['stop_year'].astype('int64')
```

```python
police.insert(2,'stop_month',0)
police['stop_month'] = police['stop_date'].str[5:7]
```

```python
police.insert(4,'stop_hour',0)
```

```python
police['stop_hour'] = police['stop_time'].str[:2]
police['stop_hour'] = police['stop_hour'].astype('int64')
```

```python
police.insert(5,'stop_daytime','')
```

```python
police.loc[(police['stop_hour'] >= 6) & (police['stop_hour'] <= 11),'stop_daytime'] = 'Morning'
police.loc[(police['stop_hour'] >= 12) & (police['stop_hour'] <= 17),'stop_daytime'] = 'Afternoon'
police.loc[(police['stop_hour'] >= 18) & (police['stop_hour'] <= 21),'stop_daytime'] = 'Evening'
police.loc[(police['stop_hour'] >= 22) | (police['stop_hour'] <= 5),'stop_daytime'] = 'Night'
```

```python
police.head().style.background_gradient(cmap='Blues')
```

```python
police = police.drop(labels=['stop_date','stop_time','driver_age_raw'],axis=1)
```

```python
police.head().style.background_gradient(cmap='Blues')
```

```python
police['driver_age'] = police['driver_age'].astype('int64')
```

```python
police['stop_month'].unique()
```

```python
police['stop_month'] = police['stop_month'].replace({'01':'Jan' ,'02':'Feb' ,'03':'Mar', '04':'Apr','05':'May','06':'June','07':'July','08':'Aug','09':'Sept','10':'Oct','11':'Nov','12':'Dec'})
```

```python
police['driver_gender'] = police['driver_gender'].replace({'M':'Male' ,'F':'Female'})
```

```python
police.head().style.background_gradient(cmap='Blues')
```

```python
for column in police:
    print("Frequancy of "  + column + " column :")
    print(police[column].value_counts())
    print("-------------------------------------------\n")
```

```python
police = police[police['stop_outcome'] != 'N/D']
police['stop_outcome'].value_counts()
```

```python
police = police[police['stop_duration'] != '2']
police = police[police['stop_duration'] != '1']
police['stop_duration'].value_counts()
```
