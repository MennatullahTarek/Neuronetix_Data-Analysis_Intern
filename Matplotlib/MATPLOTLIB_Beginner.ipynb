###  <span style="color:#fed95a;"> Q1)</span><span style="color:#32a955;">  Dual-Line Plot with Markers and Labels. </span>

```python
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.simplefilter("ignore")

# Data
months = np.array(['January', 'February', 'March', 'April', 'May', 'June'])
revenue = np.array([12000, 15000, 17000, 16000, 18000, 20000])
profit = np.array([3000, 4000, 3500, 5000, 6000, 6500])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(months, revenue, '-o', label='Revenue', marker='o', color='yellow')
plt.plot(months, profit, '--s', label='Profit', marker='s', color='green')

# Labels and Title
plt.xlabel('Months')
plt.ylabel('Amount ($)')
plt.title('Company Revenue and Profit Over Six Months')
plt.legend(loc='upper left')
plt.grid(True)

plt.show()
```

###  <span style="color:#fed95a;"> Q2)</span><span style="color:#32a955;">  Vertical Bar Chart with Annotations. </span>

```python
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.simplefilter("ignore")

# Data
grades = np.array(['Grade 1', 'Grade 2', 'Grade 3', 'Grade 4', 'Grade 5'])
books_read = np.array([25, 30, 22, 35, 28])

# Plotting
plt.figure(figsize=(8, 5))
bars = plt.bar(grades, books_read, color=['#0FFF50', '#097969', '#32a955', '#008000', '#98FB98'])

# Adding Annotations
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, yval, ha='center', va='bottom')

# Labels and Title
plt.xlabel('Grades')
plt.ylabel('Number of Books Read')
plt.title('Books Read by Students in Different Grades')

plt.show()
```

###  <span style="color:#fed95a;"> Q3)</span><span style="color:#32a955;">  Customized Pie Chart. </span>

```python
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.simplefilter("ignore")

# Data
number_of_books = np.array(['More than 20 books', '11-20 books', '6-10 books', '1-5 books', 'No books'])
readers = np.array([27, 19, 18, 32, 4])

# Plotting
plt.figure(figsize=(8, 8))
explode = [0.1 if ms == max(readers) else 0 for ms in readers]
colors = ['#b3b300', '#ffff27', '#ebeb14', '#e2e21d', '#ffff89']

plt.pie(readers, labels=number_of_books, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode)

# Title
plt.title('Books Readers')

plt.show()
```
