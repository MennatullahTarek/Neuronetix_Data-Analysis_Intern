```python
import matplotlib.pyplot as plt

# Data
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
hours_studied = [2, 3, 4, 3, 5, 2, 1]
hours_slept = [7, 6, 8, 7, 7, 9, 8]

# Plotting
fig, (ax1, ax2) = plt.subplots(2, 1, sharey=True, figsize=(8, 6))

# First plot: Hours Studied
ax1.plot(days, hours_studied, '-o', color='blue', label='Hours Studied')
ax1.set_ylabel('Hours')
ax1.set_title('Hours Studied Over a Week')
ax1.grid(True)

# Second plot: Hours Slept
ax2.plot(days, hours_slept, '--s', color='green', label='Hours Slept')
ax2.set_xlabel('Days')
ax2.set_ylabel('Hours')
ax2.set_title('Hours slept Over a Week')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Data
tasks = ['Task 1', 'Task 2', 'Task 3', 'Task 4']
team_a = [5, 3, 4, 7]
team_b = [6, 4, 3, 8]
team_c = [4, 5, 7, 6]

# Plotting
fig, ax = plt.subplots(figsize=(8, 6))

# Stacked bars
bars1 = ax.bar(tasks, team_a, color='blue', label='Team A')
bars2 = ax.bar(tasks, team_b, bottom=team_a, color='orange', label='Team B')
bars3 = ax.bar(tasks, team_c, bottom=np.array(team_a) + np.array(team_b), color='green', label='Team C')

# Adding Labels
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_y() + yval/2, yval, ha='center', va='center')

# Labels and Title
ax.set_xlabel('Tasks')
ax.set_ylabel('Performance')
ax.set_title('Team Performance in Different Tasks')
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['Housing', 'Food', 'Transportation', 'Entertainment', 'Others']
expenses = [40, 30, 10, 10, 10]

subcategories = ['Rent', 'Utilities', 'Groceries', 'Dining Out', 'Gas']
sub_expenses = [30, 10, 20, 10, 10]

# Outer Pie Chart
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=140, radius=1.2, colors=['blue', 'orange', 'green', 'red', 'purple'], explode=[.1, 0, 0, 0, 0])

# Inner Pie Chart
ax.pie(sub_expenses, labels=subcategories, autopct='%1.1f%%', startangle=140, radius=.8, colors=['lightblue', 'lightcoral', 'lightgreen', 'lightsalmon', 'lightpink'],textprops={'fontsize': 7}
)

# Title
ax.set(aspect="equal")
plt.title('Family Expenses Distribution')

plt.show()
```
